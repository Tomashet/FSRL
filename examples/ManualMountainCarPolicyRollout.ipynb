```python
import matplotlib.pyplot as plt
import numpy as np

try:
    import SafeRLBench as srb
except ImportError:
    import sys
    import os
    module_path = os.path.abspath('..')
    sys.path.append(module_path)
    import SafeRLBench as srb
```

```python
mntcar = srb.GeneralMountainCar()
roll = srb.Rollout(mntcar)

policy = lambda x: 1 + 10*x[0] + 100*x[1]
trace, i = roll.rollout(policy)

mntcar._plot()
```

```python
y = np.array([t[0] for t in [s[1] for s in trace]])
x = np.arange(0,i+1)

plt.plot(x,y)
plt.show()
```
